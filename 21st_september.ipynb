```python
import matplotlib.pyplot as plt
import numpy as np

# Create a new figure with a black background
fig, ax = plt.subplots(figsize=(8, 6))

# Set the background color to black
fig.patch.set_facecolor('black')

# Plotting the chaotic "problem" shape in red
for _ in range(10):
    t = np.linspace(0, 2*np.pi, 100)
    x = np.cos(t) + np.random.uniform(-0.5, 0.5, t.shape)
    y = np.sin(t) + np.random.uniform(-0.5, 0.5, t.shape)
    ax.plot(x - 1.5, y, color='red')

# Plotting the smooth "resolved" circle in green
circle = plt.Circle((1.5, 0), 1, color='green', fill=False, lw=2)
ax.add_artist(circle)

# Adding the text in white color
ax.text(-1.5, -2.0, 'YOUR PROBLEMS', fontsize=14, ha='center', color='white')
ax.text(1.5, -2.0, 'AFTER YOU ASK FOR HELP', fontsize=14, ha='center', color='white')

# Remove axes and set the background of the plot to black
ax.set_xlim([-3, 3])
ax.set_ylim([-2, 2])
ax.set_facecolor('black')
ax.set_aspect('equal')
ax.axis('off')

# Display the plot
plt.show()
```
